```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# 加载 Iris 数据集并转换为 Pandas DataFrame
iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# 检查数据
print(iris.head())

```

```python
# 绘制 Sepal Length 和 Sepal Width 的折线图
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'], color='blue', marker='o')
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```

```python
# 绘制 Sepal Length 和 Petal Length 的散点图
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], color='green')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```

```python
# 绘制每种类的样本数量条形图
species_count = iris['species'].value_counts()
plt.bar(species_count.index, species_count.values, color=['blue', 'green', 'red'])
plt.title("Number of Samples per Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# 绘制 Petal Length 的直方图
plt.hist(iris['petal length (cm)'], bins=20, color='purple', alpha=0.7)
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 绘制每种类样本的饼图
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', colors=['blue', 'green', 'red'])
plt.title("Sample Distribution by Species")
plt.show()
```

```python
# 设置x轴标签
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'])
plt.xlabel("Sepal Length (cm)")
plt.show()
```

```python
# 设置y轴标签
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'])
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
# 设置图表标题
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'])
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# 绘制多个系列的数据，并添加图例
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'], label="Sepal")
plt.plot(iris['petal length (cm)'], iris['petal width (cm)'], label="Petal")
plt.legend(loc="upper left")
plt.title("Sepal and Petal Length vs Width")
plt.show()
```

```python
# 保存图表为 PNG 文件
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], color='green')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.savefig('sepal_vs_petal_length.png')
plt.show()
```

```python
# 设置图表大小为 10x6 英寸
plt.figure(figsize=(10, 6))
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'], color='blue', marker='o')
plt.title("Sepal Length vs Sepal Width (Larger Figure)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```

```python
# 创建2x2的子图布局
plt.subplot(2, 2, 1)
plt.plot(iris['sepal length (cm)'], iris['sepal width (cm)'], 'r-')
plt.title("Sepal Length vs Sepal Width")

plt.subplot(2, 2, 2)
plt.plot(iris['petal length (cm)'], iris['petal width (cm)'], 'g-')
plt.title("Petal Length vs Petal Width")

plt.subplot(2, 2, 3)
plt.hist(iris['sepal length (cm)'], color='blue', alpha=0.7)
plt.title("Sepal Length Distribution")

plt.subplot(2, 2, 4)
plt.scatter(iris['petal length (cm)'], iris['sepal length (cm)'], color='purple')
plt.title("Petal Length vs Sepal Length")

plt.tight_layout()
plt.show()
```

```python
# 绘制 Petal Length 的箱线图
plt.boxplot(iris['petal length (cm)'])
plt.title("Petal Length Boxplot")
plt.show()
```

```python
# 绘制 Iris 数据集的相关矩阵热力图
import numpy as np
import seaborn as sns
corr_matrix = iris.corr()
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.title("Correlation Matrix Heatmap")
plt.show()
```
